```python
from pandas import  *
from seaborn import *
from matplotlib.pyplot import *
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import LabelEncoder
```

```python
df=read_csv('Salary_Position.csv')
df
```

```python
x=(df['Level'])
y=(df['Salary'])
```

```python
x=array(x).reshape(-1,1)
```

```python
r=LinearRegression()
r.fit(x,y)
```

```python
r.score(x,y)
```

```python
plot(x,r.predict(x),color='red')
scatter(x,y)
xlabel('Experience in years')
ylabel('Salary')
show()
```

DEGREE - 11 OR 12 IS BEST FITTED

```python
r2=PolynomialFeatures(degree=12)      # BUT why 9 is overfitting
x2=r2.fit_transform(x)   
```

```python
l=LinearRegression()
l.fit(x2,y)
```

```python
l.score(x2,y)
```

```python
plot(x,l.predict(r2.fit_transform(x)),color='red')
scatter(x,y)
xlabel('Experience in years')
ylabel('Salary')
show()
```

```python
#using linear Reg
r.predict([[5.5]])
```

```python
l.predict(r2.fit_transform([[5.5]]))
```
