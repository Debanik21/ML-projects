```python

df =read_csv('rainfallPred.csv')
df.head()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Assume 'df' contains the dataset with columns for each month's rainfall (0 or 1)

# Select the month for forecasting (change 'January' to your desired month)
selected_month = 'January'

# Extract the selected month's data
month_data = df[selected_month]

# Use ARIMA model for forecasting for the selected month
model_month = ARIMA(month_data, order=(1, 0, 0))  # Adjust the order as needed
model_fit_month = model_month.fit()  # Fit the model to the data
forecast_month = model_fit_month.forecast(steps=len(month_data))  # Forecast for the selected month

# Display the day-wise forecast for the selected month
print(f"Day-wise Forecast for '{selected_month}':")
for day, prediction in enumerate(forecast_month, start=1):
    print(f"Day {day} = {prediction}")

# Use ARIMA model for forecasting for January in the next year
# Assuming the dataset represents one year (adjust accordingly)
rainfall_data_next_year = df['January'].values  # Replace 'January' with the column representing January

model_next_year = ARIMA(rainfall_data_next_year, order=(1, 0, 0))  # Adjust the order as needed
model_fit_next_year = model_next_year.fit()  # Fit the model to the data
forecast_jan_next_year = model_fit_next_year.forecast(steps=20)  # Forecast for January next year

# Display the forecast for January next year
print("\nForecasted rainfall for January next year:")
print(forecast_jan_next_year)
```
