```python
# Importing the libraries
from pandas import  *
from seaborn import *
from matplotlib.pyplot import *
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import LabelEncoder

# Importing the dataset
df = read_csv('d3.csv')
df
```

```python
X = df.iloc[:, 0:1].values   #    x=(df[['YearsExperience']]
y = df.iloc[:, 1].values      #     y=(df[['Salary']]
```

```python


# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
r = LinearRegression()

r.fit(X, y)
```

```python
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures

r2 = PolynomialFeatures(degree=3)
x2 = r2.fit_transform(X)

```

```python
l = LinearRegression()
l.fit(x2, y)
```

```python
# Visualising the Linear Regression results
scatter(X, y, color='blue')

plot(X, r.predict(X), color='red')
title('Linear Regression')
xlabel('Exp. in Years')
ylabel('Salary')

show()
```

```python
# Visualising the Polynomial Regression results
scatter(X, y, color='blue')

plot(X, l.predict(r2.fit_transform(X)),
		color='red')
title('Polynomial Regression')
xlabel('Exp. in Years')
ylabel('Salary')
show()
```

```python
# Predicting a new result with Linear Regression
# after converting predict variable to 2D array
               #pred=Exp in years ..........(<9) accurate
l.predict(r2.fit_transform([[6.5]]))
```
